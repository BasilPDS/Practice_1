```python
import pandas as pd
```

# Предобработка
Мы открываем файл.dat  через блокнот и смотрим какие там значения. 
Типо:
1::F::1::10::48067
2::M::59::16::70075

```python
# После распаковки zip- файла каждую таблицу можно загрузить в отдельный объект DataFrame
# с помощью метода pandas.read_table:
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip'] # Даем название колонкам
# Читаем файл, разделитель в файле, пропускаем заголовки, names -  передаем сбда список названий столбцов
# engine - с помощью какого языка открываем
users = pd.read_table ('users.dat', sep='::', header=None, names=unames, engine='python')
```

```python
users[:5] # Аналог метода .head()
```

```python
# Создаем таблицу ratings

# Иницилизируем название столбцов
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ratings.dat', sep='::', header=None, names=rnames, engine='python')
ratings.head()
```

```python
# Создаем таблицу mnames

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('movies.dat', sep='::', header=None, names=mnames, engine='python')
movies.head()
```

```python
data = pd.merge(pd.merge(movies, ratings), users)
data.info()
```

##### Узнаем средний рейтинг фильмов при группировки по полу

```python
mean_ratings = data.pivot_table(values=['rating'], 
                                index=['title'],
                                columns=['gender'],
                                aggfunc='mean'
                                )
```

```python
mean_ratings
```

Хм... появились значения Nan. Это не хорошо. Надо узнать причину. Но спрема подсчитаем сколько таких значений у нас в новом датасете.

```python
mean_ratings.isna().sum()
```

Проверим фильм Zero Kelvin (Kjжrlighetens kjшtere) (1995)

```python
dNan = data[data['title'] == 'Zero Kelvin (Kjжrlighetens kjшtere) (1995)']
dNan
```

Выходит, что женщины просто не оценивали этот фильм. И судя по подсчету таких значений, женщины чаще мужчин не ставят оценки некоторым фильмам (или просто не смотрят). Может здесь будет прослеживаться связь, но вернемся к этому позже. 

Так как эти значения Nan результат отсуствия оценки в целом, можно заменить их на 0.

```python
mean_ratings_new = mean_ratings.fillna(0)
mean_ratings_new
```
